```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from datetime import datetime
from datetime import date
import os
import sys

```

```python
dir_path = './data/spot/monthly/klines/'
header = ['OpenTime', 'Open', 'High', 'Low', 'Close', 'Volume', 'CloseTime', 'QuoteAssetVolume', 'NbrTrades', 'TbuyBaseVolume', 'TbuyQuoteVolume', 'Ignore']
last_date = date(2023, 6, 29)
```

```python
symbols = os.listdir(dir_path)
print("Symbols : ", symbols)
eth_symbols = [sym for sym in symbols if 'ETH' in sym]
btc_symbols = [sym for sym in symbols if 'BTC' in sym]
eth_symbols.sort()
btc_symbols.sort()
```

```python
symbol = "ETHBTC/1h/"
dir_sym = dir_path + symbol

months = os.listdir(dir_sym)
months.sort()
print(months)

header = ['OpenTime', 'Open', 'High', 'Low', 'Close', 'Volume', 'CloseTime', 'QuoteAssetVolume', 'NbrTrades', 'TbuyBaseVolume', 'TbuyQuoteVolume', 'Ignore']
df_eth = pd.read_csv(dir_sym+months[-1], names=header)
df_eth.set_index('OpenTime', inplace=True)
df_eth
```

## Load data


```python
dict_symbols = {}

for symbol in symbols :
    print(symbol, end = ' -- ')
    symbol_freq = symbol+"/1h/"
    
    dir_sym = dir_path + symbol_freq
    
    try : 
        months = os.listdir(dir_sym)
        months.sort()

        df_temp = None 
        for m in months : 
            temp = pd.read_csv(dir_sym+m, names=header)
            temp['OpenTime'] = pd.to_datetime(temp['OpenTime']*10**6)
            temp.set_index('OpenTime', inplace=True)
            df_temp = pd.concat([df_temp, temp])
    except : 
        print(dir_sym, )
            
    
    dict_symbols[symbol] = df_temp
```

## Only keep still existing cryptos

```python
keys = list(dict_symbols.keys())
eth_keys = [sym for sym in keys if 'ETH' in sym]
eth_keep = []
len(eth_keys)
for k in eth_keys: 
    last_date_symbol = dict_symbols[k].index[-1]
    if(last_date_symbol > last_date): 
        eth_keep.append(k)
        print(k, last_date_symbol, dict_symbols[k].index[0], len(dict_symbols[k].index))
```

```python
keys = list(dict_symbols.keys())
eth_keys = [sym for sym in keys if 'ETH' in sym]
len(eth_keys)

intersection = dict_symbols[eth_keep[0]].index

for k in eth_keep: 
    print(k)
    symbol = (dict_symbols[k])
    plt.plot(symbol.Open*dict_symbols['ETHUSDT'].Open)
    plt.show()
    intersection = list(set(intersection) & set(symbol.index))
print(len(intersection))
```

```python
last_df = {}
open_df = {}
# open_df['time'] = intersection

for k in eth_keep: 
    print(k, end =' -- ')
    symbol = (dict_symbols[k])
    selection = symbol.index.isin(intersection)
    last_df[k] = symbol[selection]
    open_df[k] = last_df[k].Open

df = pd.DataFrame(open_df)
df.head()
    
```

```python
def featureQQ(x, y, col=""): 
    feature = x.shift(1)[1:]
    obj = y[1:]
    print(col, np.corrcoef(feature, obj)[0,1])

def feature_finder(X, y):
    for col in X.columns : 
        featureQQ(X[col], y, col)
        
feature_finder(df_returns, df_returns['ETHBTC'])
    
```
